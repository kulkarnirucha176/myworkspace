### Metric 1: Weeks at No.1:
#### This metric refers to the No.1 Rank of a tennis player for consecutive weeks in a year. The highest number of weeks at no.1 is an indicator of consistency of performance for a tennis player. Consistency is derived from the position of the player in the ATP Rankings – which is based on a points system.

#### Rebuttal: The “Weeks at No.1” position indicates that the player (in Men’s singles) could consistently beat his competitors and maintain the top position every week.  This is a difficult feat to achieve since there are multiple matches held all over the world and everyone is of course, playing to win!

#### Success factor/KPI: Highest number of Weeks at No. 1


### Metric 2: Years at No.1

#### This metric refers to the No.1 Rank of a tennis player at the end of a year. The highest number of years at no.1 does not necessarily indicate consistency of success. While it is commendable that the player in question has the capability to regain the top position (in some cases), it does not necessarily imply consistency. In the game of tennis, it is the consistency factor that is highly valued. 

#### Warrant: There have been situations in the past where a player has dominated Tennis for a year, not play for 1-2 years and regain the number 1 position in the next year. 

#### Success factor/KPI: Highest number of years at the top position at the end of each year

### Metric 3: Number of Grand slam titles
#### This metric is all about the number of Grand slam titles the player has won. In this case, it would be interesting to see who has the highest number of titles and who the closest competitor is. The players need to play on all four surfaces of the Grand Slam tournament. 

#### Rebuttal: It is obvious that this would be the most important factor to judge the greatness of a player.

#### Success factor/KPI: Highest number of Grand Slam titles

### Metric 4: Grand Slam Surface expertise
#### This metric is all about the number of Grand slam titles the player has won on a surface. In this case, it would be interesting to see who has the highest number of titles and who the closest competitor is on that surface. Would it really matter if a player is the greatest in a surface but is not the Grand Slam leader? 

#### Rebuttal: As all the players need to play on all four surfaces of the Grand Slam tournament, it is fair to compare the success when a player is great on all surfaces and not just one. Just the way students are graded on all the subjects before declaring who achieved 4.0.

#### Success factor/KPI: Highest number of Grand Slam titles on a surface
